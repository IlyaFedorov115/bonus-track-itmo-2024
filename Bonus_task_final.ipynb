#### приведение имен к одному виду

```python
def match_names(row, df1):
    for name in df1['name']:
        if row['name'] in name:
            return name
    return None
```

```python
data2['name'] = data2.apply(match_names, axis=1, df1=data1)
```

```python
data2 = data2.dropna(subset=['name'])
len(data2)
```

### Нормализация длины котов

```python
df1_length_list = list(df_merged['length'])
df1_length_list = [elem.lower() for elem in df1_length_list]
df1_length_list
```

```python
df1_length_list = [str.replace('short', 'small') for str in df1_length_list]
df1_length_list = [str.replace('long', 'large') for str in df1_length_list]
df1_length_list = [str.replace('inches', '') for str in df1_length_list]
df1_length_list = [str.strip() for str in df1_length_list]

df1_length_list
```

```python
df1_length_list = [str.split(',', 1)[0] for str in df1_length_list]
df1_length_list = [str.strip() for str in df1_length_list]

df1_length_list
```

```python
import random
```

```python
def get_random_length(txt_format, single=False):
  if type(txt_format) is str and single:
    if txt_format.lower() == 'small':
      return 8 + random.randint(1,5)
    elif txt_format.lower() == 'medium':
      return 13 + random.randint(1,5)
    elif txt_format.lower() == 'large':
      return 19 + random.randint(1,7)
    else:
      raise Exception("Unknown length name '" + txt_format + "' !")

  if type(txt_format) is str:
    if txt_format.lower() == 'small':
      left = 8 + random.randint(1,2)
      right = left + random.randint(1,3)
      return left, right
    elif txt_format.lower() == 'medium':
      left = 13 + random.randint(1,2)
      right = left + random.randint(1,3)
      return left, right
    elif txt_format.lower() == 'large':
      left = 19 + random.randint(1,3)
      right = left + random.randint(1,4)
      return left, right
    else:
      raise Exception("Unknown length name '" + txt_format + "' !")

  if type(txt_format) is list:
    if len(txt_format) == 1:
      return get_random_length(txt_format[0])
    elif len(txt_format) != 2:
      raise Exception("Length of list names for length must be 1 or 2, but get " + str(len(txt_format)))
    elif txt_format[0].isdigit() and txt_format[1].isdigit():
      return int(txt_format[0]), int(txt_format[1])
    else:
      return get_random_length(txt_format[0], True), get_random_length(txt_format[1], True)
  else:
    raise Exception("Type must be str or list!")
```
